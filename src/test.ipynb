```python
import pandas as pd
import numpy as np
```

```python
n = 100
get_price_series = lambda n=n, p=100: (
    pd.Series(
        np.random.normal(loc=0, scale=0.01, size=n)
        )
        .add(1)
        .cumprod()
        .multiply(p)
)

df = pd.DataFrame({
    "stock_a": get_price_series(),
    "stock_b": get_price_series(),
    "stock_c": get_price_series(),
})
df
```

```python
# history() returns a DataFrame with a MultiIndex (symbol, time)
# and columns: close, high, low, open, volume

symbols = ["AAPL", "MSFT", "GOOGL", "AMZN", "NVDA"]
dates = pd.date_range("2024-01-01", periods=5, freq="B")

idx = pd.MultiIndex.from_product([symbols, dates], names=["symbol", "time"])

np.random.seed(42)
n = len(idx)
raw = pd.DataFrame({
    "close":  get_price_series(n),
    "high":   get_price_series(n),
    "low":    get_price_series(n),
    "open":   get_price_series(n),
    "volume": np.random.uniform(1e6, 1e8, n),
}, )
raw.index = idx

raw
```
